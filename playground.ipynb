```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
datasets = ["ECU-IoHT", "WUSTL-EHMS-2020"]

# Total records in each dataset
total_records = [111207, 16318]

# Attack vs Normal counts for each dataset
# attack_counts = [87754, 11, 2046]
# normal_counts = [23453, 8, 14272]
attack_counts = [87754, 2046]
normal_counts = [23453, 14272]

# --- Create subplots ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# =====================
# Left: Total Records
# =====================
bars1 = axes[0].bar(datasets, total_records, color=[
                    "steelblue", "lightsteelblue", "darkslategray"])
axes[0].set_title("Total Records in Each Dataset")
axes[0].set_ylabel("Number of Records")
axes[0].set_axisbelow(True)
axes[0].yaxis.grid(True, linestyle='--', alpha=0.7)

# Add labels on top
for bar, count in zip(bars1, total_records):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                 f"{count:,}", ha='center', fontsize=10, fontweight="bold")

# =====================
# Right: Attack vs Normal
# =====================
x = np.arange(len(datasets))  # dataset positions
width = 0.35  # bar width

bars2 = axes[1].bar(x - width/2, attack_counts, width,
                    label="Attack", color="lightcoral")
bars3 = axes[1].bar(x + width/2, normal_counts, width,
                    label="Normal", color="lightgreen")

axes[1].set_title("Attack vs Normal Distribution")
axes[1].set_xticks(x)
axes[1].set_xticklabels(datasets)
axes[1].set_ylabel("Count")
axes[1].legend(title="Traffic Type")

axes[1].set_axisbelow(True)
axes[1].yaxis.grid(True, linestyle='--', alpha=0.7)

# Add labels on top of bars
for bar, count in zip(bars2, attack_counts):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                 f"{count:,}", ha='center', fontsize=9, color="coral", fontweight="bold")

for bar, count in zip(bars3, normal_counts):
    axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
                 f"{count:,}", ha='center', fontsize=9, color="green", fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the table
algorithms = [
    "Random Forest", "XGBoost", "SVM",
    "LightGBM", "CatBoost", "KNN", "MLP",
    "Logistic Regression", "AdaBoost", "Gradient Boosting", "Naive Bayes"
]
original_best = [0.9623, 0.9655, 0.9356, 0.9647, 0.9662,
                 0.956, 0.8923, 0.7628, 0.958, 0.9651, 0.7692]
augmented_best = [0.991, 0.9914, 0.9786, 0.9918, 0.9916,
                  0.971, 0.9804, 0.9172, 0.9865, 0.9915, 0.9152]

# Custom colors for higher contrast (can be expanded/changed)
custom_colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd",
    "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf", "#ff1493"
]

plt.figure(figsize=(10, 6))
for i, algo in enumerate(algorithms):
    plt.plot(["Original Best", "Augmented Best"],
             [original_best[i], augmented_best[i]],
             marker='o', color=custom_colors[i % len(custom_colors)],
             label=algo)

plt.xlabel("Dataset Type", fontweight='bold')
plt.ylabel("Score", fontweight='bold')
plt.title("Overall Comparison of Existing and Augmented Datasets Trained Models",
          fontsize=14)
plt.ylim(0.7, 1.05)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.legend(title="Algorithms", loc='lower center',
           bbox_to_anchor=(0.5, -0.3), ncol=6)
plt.tight_layout()
plt.show()
```
